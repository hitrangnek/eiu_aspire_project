```python
import pandas as pd

# Assuming 'df' is your DataFrame
# First, let's check the columns available in the DataFrame
print(df1.columns)

# Optionally strip spaces in column names if there are any issues with extra spaces
df1.columns = df1.columns.str.strip()

# Select the desired columns, if they exist in the DataFrame
if 'Id_company' in df1.columns and 'Phone' in df1.columns:
    data = df1[['Id_company', 'Phone']]
    print(data.head())
else:
    print("The columns 'Id_company' or 'Phone' do not exist in the DataFrame.")
```

1. Modify the phone number to replace the country code +84 with 0.
2. Convert dots (.) in the phone numbers
3. Remove any remaining spaces to normalize the phone numbers

```python
data['Phone'] = data['Phone'].str.replace(r'\+84', '0', regex=True)
data['Phone'] = data['Phone'].str.replace(r'\.', '', regex=True)
data['Phone'] = data['Phone'].str.replace(r'[() ]', '', regex=True)
data['Phone'] = data['Phone'].str.replace(r'^(\+84|0084|84)', '0', regex=True)
```

4. Caculate in columns ( <13 characters) and remove "-" with "" for the first .

Normally, Vietnamese phone number have the length from 10-11. And some of rows in the dataset have 2 phone numbers, based on that => select thresolf of 13 to remove '-' , some of case 2 phone number connect with '-', difficult to preprocessing if choose this case.*


```python
def process_phone_numbers(data):
    # def function to preprocessing number_phone
    def process_number(phone):
        # Loại bỏ dấu "-" và đếm số ký tự #Remove "-"
        phone = phone.replace('-', '')

        # Check the length of character <=13
        if len(phone) <= 13:
            return phone
        return phone  # If longer is longer 13, keep

    # Apply the function
    data['Processed_Phone'] = data['Phone'].apply(process_number)

    return data
```

```python
data['Phone'] = data['Phone'].str.replace(r'\.', '', regex=True)  # Remove "."
data['Phone'] = data['Phone'].str.replace(r'\s', '', regex=True)  # Remove empty space
```

```python
def remove_non_numeric_characters(data):
    # Check the column names in the DataFrame to ensure the column names are correct
    print(data.columns)  # Display the column names in the DataFrame

    # Function to process phone numbers
    def process_number(phone):
        # Replace all non-numeric characters (letters, commas, spaces, etc.) with a '*'."
        phone = ''.join([i if i.isdigit() else '*' for i in phone])  # Chỉ thay ký tự không phải số bằng "*"
        return phone

    # Apply the processing function to the 'Phone' column
    data['Phone'] = data['Phone'].apply(process_number)

    # Rename the 'Phone' column to 'phone_number_1
    data.rename(columns={'Phone': 'phone_number_1'}, inplace=True)

    return data

processed_data = remove_non_numeric_characters(data)

# Display the results
print(processed_data[['phone_number_1']].head())
```

```python
def process_phone_numbers_and_count(data):
    # Count the number of characters in the 'phone_number_1' column.
    char_count = data['phone_number_1'].apply(lambda x: pd.Series(list(x))).stack().value_counts()

    # Def function for preprocessing phone_number
    def process_number(phone):
        if len(phone) > 15:
            # "Cut the part after the '*' and store it in phone_number_2.
            phone_number_1 = phone.split('*')[0]
            phone_number_2 = phone.split('*')[1] if len(phone.split('*')) > 1 else None
            return phone_number_1, phone_number_2
        else:
            # If the length is less than or equal to 15, keep it unchanged
            return phone, None

    # Apply the processing function to the 'phone_number_1' column
    data[['phone_number_1', 'phone_number_2']] = data['phone_number_1'].apply(lambda x: pd.Series(process_number(x)))

    return data, char_count

processed_data = data
processed_data, char_count = process_phone_numbers_and_count(processed_data)
```
